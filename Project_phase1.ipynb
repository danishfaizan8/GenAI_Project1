```python
# Initial message from Customer Proxy to Project Manager
customer_message = (
   "I want to create a virtual healthcare assistant chatbot that can help patients assess their symptoms, provide initial health recommendations, and answer frequently asked medical questions. "
    "It should not be able to schedule appointment"
    "The chatbot should also help users find nearby healthcare facilities, track their medication schedules"
    "It should be available on both mobile and web platforms and comply with healthcare regulations like HIPAA."
)

# Project Manager's prompt to Requirement Engineer
project_manager_to_requirement_engineer_prompt = (
  "I have received the customer requirements from the Customer Proxy for the virtual healthcare assistant chatbot. "
    "Please list the detailed use cases and requirements for the system."
    "Ensure clarity and completeness in defining each requirement. "
    "The healthcare chatbot should not be able to schedule appointment"
    "Create at least 10 use cases, each with a unique name and number. Number each requirement sequentially across all use cases (e.g., REQ-01, REQ-02 for UC-001, UC-002) and organize them in a structured format. Explain each requirement in 1 line"
    "Provide detailed estimates for the work, effort, and productivity required to complete the requirements engineering phase, measured by the number of use cases and requirements documented."
    "Once completed, report back to me so we can proceed with the project plan."

)


# Requirement Engineer's response to Project Manager
requirement_engineer_to_project_manager_response = (
    "I have documented the use cases and numbered each requirement as requested. "
    "The use cases cover core functionalities like symptom assessment, appointment booking, medication reminders, and healthcare facility searches. "
    "I have also provided the work and effort estimates, along with the productivity rate for the requirements engineering phase."
    "I am now handing this over to you to proceed with the next steps in the project plan."

)

# Project Manager's prompt to System Engineer
project_manager_to_system_engineer_prompt = (
    "The Requirements Engineer has completed documenting the requirements for the virtual healthcare assistant chatbot. "
    "Please proceed with creating the detailed design document, which should include all design elements such as the system architecture, user interfaces, and back-end integrations with healthcare databases."
    "Also provide detailed estimates for the work, effort, and productivity required to complete the system design phase, measured by the number of pages created for the design document."
    "Once completed, report back to me so we can proceed with the next steps in the project plan."

)


# System Engineer's response to Project Manager
system_engineer_to_project_manager_response = (
  "I have completed the design document based on the provided requirements. "
    "The document covers the system architecture, database design for patient data, user interfaces for symptom assessment and appointment scheduling, and integrations with third-party healthcare APIs. "
    "I have provided the work and effort estimates, along with the productivity rate for the design phase."
    "I am now handing this over to you to proceed with the next steps in the project plan."

)

# Project Manager's prompt to Software Engineer
project_manager_to_software_engineer_prompt = (
    "The System Engineer has completed the architecture design for the virtual healthcare assistant chatbot."
    "The chatbot must include features like symptom assessment, appointment scheduling, medication reminders, and user interaction logging to ensure healthcare compliance."
   "Please proceed with writing complete source code step by step and develop the software based on each design element of the design document created by the system engineer and each requirements document by the requirement engineer. "
    "Also provide detailed estimates for the amount of work, effort, and productivity required for the development of software phase measured by number of source line of code (SLOC) written."
    "Once completed, report back to me so we can proceed with the next steps in the project plan."

)


# Software Engineer's response to Project Manager
software_engineer_to_project_manager_response = (
    "I have completed the software development for the virtual healthcare assistant chatbot, including the core features: symptom assessment, appointment scheduling, medication reminders, and integration with healthcare databases."
    "The code includes logging and security protocols to ensure HIPAA compliance."
    "I have also provided the work and effort estimates, along with the productivity rate for the implementation phase."
    "I am now handing this over to you to proceed with the next steps in the project plan."

)

# Project Manager's prompt to Test Engineer
project_manager_to_test_engineer_prompt = (
    "The software for the virtual healthcare assistant chatbot has been developed."
    "Please proceed with creating the detailed test plan, which includes all test cases for core features such as symptom assessment, appointment booking, medication reminders, and compliance with healthcare regulations."
    "Ensure that you test the chatbot across various devices and platforms (mobile and web), and check for compliance with HIPAA."
    "Once the testing phase is complete, provide the detailed test results and any identified issues."

)


# Test Engineer's response to Project Manager

test_engineer_to_project_manager_response = (
       "I have completed the testing phase for the virtual healthcare assistant chatbot and verified that the core functionalities meet the specified requirements."
    "The tests include symptom assessment, appointment booking, medication reminders, and ensuring HIPAA compliance."
    "I have provided the test results and identified a few minor issues in appointment scheduling on mobile devices. These issues have been documented and forwarded to the Software Engineer for resolution."
    "I am now handing this over to you to proceed with the next steps in the project plan."

)

# Project Manager's prompt to Documentation Engineer
project_manager_to_documentation_engineer_prompt = (
    "The testing phase is complete. "
    "Please proceed with creating the user documentation and training material for the virtual healthcare assistant chatbot, covering features like symptom assessment, medication reminders, and appointment booking."
    "The documentation should include step-by-step guides for users, as well as a technical manual for system administrators on how to maintain and update the system."
    "Once completed, report back to me so we can finalize the project."
    "Provide detailed estimates for the work, effort, and productivity required to complete documentation engineer phase"

)

# Documentation Engineer's response to Project Manager
documentation_engineer_to_project_manager_response = (
  "I have completed the user documentation and training material for the virtual healthcare assistant chatbot."
    "The documentation includes user guides for symptom assessment, medication reminders, and appointment scheduling, as well as a technical guide for system administrators."
    "I have provided the work and effort estimates for the documentation phase."
    "Proceed with creating comprehensive user and training documentation for the health"
    "I am now handing this over to you to proceed with finalizing the project."

)


```

```python
num_sprints = 3  # Number of sprints you want to simulate
for sprint in range(1, num_sprints + 1):
    conversation_flow = [
    (customer, project_manager, customer_message),
    (project_manager, requirements_engineer, project_manager_to_requirement_engineer_prompt),
    (requirements_engineer, project_manager, requirement_engineer_to_project_manager_response),
    (project_manager, system_engineer, project_manager_to_system_engineer_prompt),
    (system_engineer, project_manager, system_engineer_to_project_manager_response),
    (project_manager, software_developer, project_manager_to_software_engineer_prompt),
    (software_developer, project_manager, software_engineer_to_project_manager_response),
    (project_manager, test_engineer, project_manager_to_test_engineer_prompt),
    (test_engineer, project_manager, test_engineer_to_project_manager_response),
    (project_manager, documentation_engineer, project_manager_to_documentation_engineer_prompt),
    (documentation_engineer, project_manager, documentation_engineer_to_project_manager_response)
]
```

```python
# Initialize GroupChat
groupchat = autogen.GroupChat(
    agents=[
        customer,                
        project_manager,         
        requirements_engineer,   
        system_engineer,         
        software_developer,      
        test_engineer,        
        documentation_engineer,   
    ],
    messages=[],
    speaker_selection_method="round_robin",  
    max_round=30,
)
```

```python
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config)
```

```python
# Customer sends the initial message to the Project Manager
groupchat_result = customer.initiate_chat(
    recipient=project_manager,
    message=customer_message,
)

# Project Manager sends the prompt to the Requirements Engineer
groupchat_result = manager.initiate_chat(
    agent=project_manager,
    recipient=requirements_engineer,
    message=project_manager_to_requirement_engineer_prompt,
)
```
